```python
from pyswip import Prolog

prolog = Prolog()

prolog.assertz("character(rama)")
prolog.assertz("character(sita)")
prolog.assertz("character(lakshman)")
prolog.assertz("character(hanuman)")
prolog.assertz("character(ravana)")

prolog.assertz("relationship(rama, sita, husband)")
prolog.assertz("relationship(sita, rama, wife)")
prolog.assertz("relationship(rama, lakshman, brother)")
prolog.assertz("relationship(lakshman, rama, brother)")

prolog.assertz("event(abduction_of_sita, ravana, sita)")
prolog.assertz("event(search_for_sita, rama, lakshman)")
prolog.assertz("event(bridge_building, hanuman)")

prolog.assertz("location(ayodhya)")
prolog.assertz("location(lanka)")
prolog.assertz("location(forest)")

prolog.assertz("at(rama, ayodhya)")
prolog.assertz("at(sita, ayodhya)")
prolog.assertz("at(lakshman, ayodhya)")
prolog.assertz("at(ravana, lanka)")

prolog.assertz("hero(X) :- character(X), (X == rama ; X == lakshman ; X == hanuman)")
prolog.assertz("villain(X) :- character(X), X == ravana")
prolog.assertz("abducted_by(sita, ravana)")

print("Who is Rama's wife?")
for soln in prolog.query("relationship(rama, X, husband)"):
    print(soln["X"])

print("\nWho is Sita's husband?")
for soln in prolog.query("relationship(sita, X, wife)"):
    print(soln["X"])

print("\nWho are Rama's brothers?")
for soln in prolog.query("relationship(rama, X, brother)"):
    print(soln["X"])

print("\nWho abducted Sita?")
for soln in prolog.query("abducted_by(sita, X)"):
    print(soln["X"])

print("\nFind out who is a hero")
for soln in prolog.query("hero(X)"):
    print(soln["X"])

print("\nFind out who is a villain")
for soln in prolog.query("villain(X)"):
    print(soln["X"])
```
